```
F=open("test1.txt","r+")
F=F.readlines()
#print(F)
G=[]
L=[]
R=[]
for i in F:
   G.append(i.split(" "))
for i in G:
   L=i
for i in L:
  R.append(int(i))
print(R)
```

```
M=[]
c=0
for i in range(len(L)):
  c=c+1
  M.append(c)
print(M)
```

```
import numpy as np
import matplotlib.pyplot as plt
plt.plot(M,R)
```

```
P=[]
for i in range (1,9999):
  if(int(R[i])>0):
    if (R[i-1]<R[i] and R[i]>R[i+1]):
      P.append(R[i])
print(P)
```

```
k=0
K=[]
for i in P:
  k=k+1
  K.append(k) 
print(K)
```

```
plt.plot(K,P)
```

```
Rpeaks=[]
for i in range(len(P)):
  if P[i]>800:
    Rpeaks.append(P[i])
    print(i)
```

```
sma=[]
for i in range(2,len(P)):
    j=(P[i-2]+P[i-1]+P[i])/3
    sma.append(j)
v=0
V=[]
for i in sma:
  v=v+1
  V.append(v)
plt.plot(V,sma)
```

```
peaks=[]
b=0
ipeaks=[]
 
for i in range (1,701):
    if (sma[i-1]<sma[i] and sma[i]>sma[i+1]):
      peaks.append(sma[i])
for i in peaks:
  b=b+1
  ipeaks.append(b)
plt.plot(ipeaks,peaks)
print(peaks)
```

```
peakpos=[]
rpos=[]
for i in range(len(peaks)):
  if peaks[i]>=100 and peaks[i]<300:
    peakpos.append(i)
for i in range(len(peaks)):
  if peaks[i]>300:
    rpos.append(i)
print(rpos)
len(peakpos)
```

```
import numpy as np
!pip install heartpy 
import heartpy as hp
```

```
R_=hp.filtering.filter_signal(R, 0.6, 1000, order=2, filtertype='highpass', return_top=False)
plt.plot(M,R_)
plt.plot(M,R)
```

```
R__=hp.filtering.filter_signal(R_, 50, 1000, order=2, filtertype='notch', return_top=False)
plt.plot(M,R_)
plt.plot(M,R)
plt.plot(M,R__)
```

```
R___=hp.filtering.filter_signal(R__, 100, 1000, order=2, filtertype='lowpass', return_top=False)
 
plt.plot(M,R___)
```

```
P_=[]
peaksr_=[]
for i in range (1,9999):
  if(int(R___[i])>0):
    if (R___[i-1]<R___[i] and R___[i]>R___[i+1]):
      if R___[i]>800:
        P_.append(R___[i])
        peaksr_.append(i)
 
len(P_)
```

```
S_=hp.filtering.hampel_filter(R___, filtsize = 6)
plt.plot(M,S_)
```

```
S__=hp.filtering.smooth_signal(S_, 1000, window_length=None, polyorder=3)
plt.plot(M,S__)
len(S__)
```

```
peaksr=[]
peaksrl=[]
for i in range (1,9999):
  if (S__[i])>0:
    if (S__[i-1]<S__[i] and S__[i]>S__[i+1]):
      if S__[i]>200:
        peaksr.append(S__[i])
        peaksrl.append(i)
 
print(peaksrl)
```

```
peakspt=[]
peakspt_=[]
for i in range(0,9999):
   if S__[i]>S__[i-1] and S__[i]>S__[i+1]:
      if S__[i]<200 and S__[i]>0:
         peakspt.append(S__[i])
         peakspt_.append(i)
print(peakspt_)
```

```
peaksall=[]
peaksall_=[]
for i in range (1,9999):
  if (S__[i])>0:
    if (S__[i-1]<S__[i] and S__[i]>S__[i+1]):
      if (S__[i]>200) or (S__[i]<200 and S__[i]>0):
        peaksall.append(S__[i])
        peaksall_.append(i)
 
len(peaksall)
```

```
plt.plot(M,S__)
plt.plot(peakspt_,peakspt)
plt.plot(peaksr_,P_)
plt.plot(peaksr_, P_, 'o', color='green');
```

```
peaksp=[]
peakst=[]
peaksp_=[]
peakst_=[]
for i in range(len(peaksall)):
  if peaksall[i] in peaksr:
    peakst.append(peaksall[i+1])
    peakst_.append(peaksall_[i+1])
    peaksp.append(peaksall[i-1])
    peaksp_.append(peaksall_[i-1])
plt.plot(peaksall_,peaksall)
plt.plot(peaksp_,peaksp)
plt.plot(peakst_,peakst)
```
